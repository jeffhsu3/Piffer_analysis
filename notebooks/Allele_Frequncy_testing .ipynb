```python
url_1000G_phase3_base = '../../1KG_proj/'
url_1000G_phase3_prefix = 'ALL.chr'
url_1000G_phase3_suffix = '.phase3_shapeit2_mvncall_integrated_v5.20130502.genotypes.vcf.gz'
populations_1000G = pd.read_csv(url_1000G_phase3_base + 'integrated_call_samples_v3.20130502.ALL.panel', 
                                sep = "\t", index_col = 0)
```

```python
def parse_header(vcf_header):
    for i in vcf_header:
        if i[0:2] == '##':
            pass
        else: 
            out = i.split("\t")
            out[0].lstrip("#")
            return(out)

def convert_vcf_code(genotype_string, reverse = False):
    """ Convert vcf coded genotypes to numerics
    """
    
    try:
        if reverse:
            out = abs((int(genotype_string[0]) - 1) + (int(genotype_string[2])-1))
        else:
            out = int(genotype_string[0]) + int(genotype_string[2]) 
    except IndexError:
        out = np.nan
    
    return(out)


def calculate_allele_frequency(genotypes):
    """ Calculates the allele frequency to the allele that is coded as the 1.
    """
    return(np.sum(genotypes)/float(2*len(genotypes)))


   
```

```python
def write_allele_frequencies(chromosome, populations_1000G):
    """ Convenience function for calculating and writing out allele frequencies given a popoulation
    """
    global url_1000G_phase3_base
    global url_1000G_phase3_prefix 
    global url_1000G_phase3_suffix
    filename = url_1000G_phase3_base + url_1000G_phase3_prefix + str(chromosome) + url_1000G_phase3_suffix 
    BASE_out = '../../1KG_proj/allele_frequncies/'
    f = gzip.open(filename, 'rb')
    tabix = pysam.Tabixfile(filename)
    header = parse_header(tabix.header)
    pop_ids = populations_1000G.ix[header[9:], 'pop']
    meta = ['chr', 'start', 'rsID']
    pop_keys =  list(set(pop_ids.values))
    pop_keys.sort()
    meta.extend(pop_keys)
    #df = pd.DataFrame(columns=meta)
    debug = 0
    for i in f:
        if i[0:2] == '##': pass
        else: break
            
    with open(BASE_out + 'AF' + str(chromosome) + '.txt', 'w') as out_file:
        out_file.write("\t".join(meta) + "\n")
        #tabix_iter = tabix.fetch('22') 
        ids_index = pd.Index(header[9:])
        for i in f:
            i = i.split("\t")
            if i[2] == ".":
                rsid = i[0] + "_" + i[1] + "_" + i[4] 
            else:
                rsid = i[2]
            try:
                genotypes = pd.Series([convert_vcf_code(j, reverse = False) for j in i[9:]], index=ids_index)
            except ValueError:
                print(i[9:])
                
            out = genotypes.groupby(pop_ids).aggregate(calculate_allele_frequency)
            out = out.round(decimals=5)
            bleh = [i[0], i[1], rsid]
            bleh.extend(out[pop_keys].values.astype('|S7'))
            #df.loc[rsid] = bleh
            #print(out.values)
            out_file.write("\t".join(bleh) + "\n")
            #out_file.write("\t".join(out[pop_keys].values.astype('|S7')) )
            """
            debug +=1
            if debug > 2000:
                #return(df)
                break
            """
     
   
```
