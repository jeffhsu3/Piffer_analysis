# Polygenic Selection for Atrial Fibrillation

A replicaiton of Piffer(2013), Piffer(2014), and Davide Piffer and Gilfoyle (2014) work on polygenic selection on education-attainment and cognitive capability.  This first part contains the calculation for the PCA's based on the Reitner papers.  The latter parts require that you have available all of the 1000 Genomes Data as well as some other publicly avaiable population data.

```python
# Move these utility functions to a separate file or more usefule in showing what is being done?  
def parse_header(vcf_header):
    for i in vcf_header:
        if i[0:2] == '##':
            pass
        else: 
            out = i.split("\t")
            out[0].lstrip("#")
            return(out)


   
def convert_vcf_code(genotype_string, reverse = False):
    """ Convert vcf coded genotypes to numerics
    """
    
    try:
        if reverse:
            out = abs((int(genotype_string[0]) - 1) + (int(genotype_string[2])-1))
        else:
            out = int(genotype_string[0]) + int(genotype_string[2]) 
    except IndexError:
        out = np.nan
    
    return(out)


def calculate_allele_frequency(genotypes):
    """ Calculates the allele frequency to the allele that is coded as the 1.
    """
    return(np.sum(genotypes)/float(2*len(genotypes)))



def grab_allele_frequencies_by_pop(row, pop_panel, prefix = url_1000G_phase3_base + url_1000G_phase3_prefix, 
                                   suffix = url_1000G_phase3_suffix):
    """ Grabs the allele frequency by population from 1000 genomes.  Meant to be applied on a pandas DataFrame.  
    
    Arguments 
    ---------
    row - A pandas series of SNPs that you are interested in.  Object must contain 'Chrom' and 'Pos' attributes
    
    Returns
    -------
    A pandas DataFrame of the coded allele frequencies for the given SNPs
    """
    tabix = pysam.Tabixfile(prefix + str(row['Chrom']) + suffix)
    header = parse_header(tabix.header)
    pop_ids = pop_panel.ix[header[9:], 'pop']
    tabix_iter = tabix.fetch(str(row['Chrom']), row['Pos'] - 1, row['Pos'])
    for i in tabix_iter:
        i = i.split("\t")
        if i[2] == row.name:
            if i[4] == row.CodedAllele:
                recode = False
            else:
                recode = True
            genotypes = pd.Series([convert_vcf_code(j, reverse=recode) for j in i[9:]], 
                                  index = pd.Index(header[9:]))
            return(genotypes.groupby(pop_ids).aggregate(calculate_allele_frequency))
        else: pass
```

```python
freqs = AF_GWAS.apply(grab_allele_frequencies_by_pop, axis=1, args=(populations_1000G,))
```

```python
freqs
```

```python
# Fitting the PCA to the allele frequencies
pca = PCA()
pca.fit(freqs.T)
print(pca.explained_variance_ratio_) 
freqs_reduced = pca.fit_transform(freqs.T)
freqs_reduced = pd.DataFrame(freqs_reduced.T, 
                             index=pd.Index(['PCA' + str(i+1) for i in range(freqs_reduced.shape[1])]),
                             columns = freqs.columns)
```

```python
# Grabbing the correlation for each SNP with PCA1
PCA1_corr = np.zeros(freqs.shape[0], dtype=np.double)
for i in range(freqs.shape[0]):
    PCA1_corr[i] = freqs.ix[i,:].corr(freqs_reduced.ix['PCA1', :])
PCA1_corr = pd.DataFrame(PCA1_corr, index=freqs.index, columns=['PCA1 loadings'])
PCA1_corr['CodedAllele'] = AF_GWAS.CodedAllele
PCA1_corr
```
